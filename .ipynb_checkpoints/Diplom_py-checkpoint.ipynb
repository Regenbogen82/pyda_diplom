**2. Рассчитать основные статистики для переменных (среднее,медиана,мода,мин/макс,сред.отклонение).**

Так как метод describe показывает не все необходимые нам статистики, сделаем отдельную обработку. Для этого создадим отдельный dataframe, где будут учитываться только те столбцы, в которых приведены переменные для статистик.

```python
hr.number_project.unique()
```

```python
hr_var = hr[['satisfaction_level', 'last_evaluation', 'average_montly_hours', 'time_spend_company',
   'Work_accident', 'left', 'salary_cat', 'sales_cat']]
```

```python
i = 0
dict_stats = {'variables':[], 'среднее': [], 'медиана': [], 'минимум': [], 'максимум': [], 'стандартное отклонение': [], 'мода': []}
while i < 8:
    dict_stats['variables'].append(hr_var.columns[i])
    dict_stats['среднее'].append(statistics.mean(hr_var[hr_var.columns[i]])) 
    dict_stats['медиана'].append(statistics.median(hr_var[hr_var.columns[i]]))
    dict_stats['минимум'].append(min(hr_var[hr_var.columns[i]]))
    dict_stats['максимум'].append(max(hr_var[hr_var.columns[i]]))
    dict_stats['стандартное отклонение'].append(statistics.pstdev(hr_var[hr_var.columns[i]]))
    try:
        dict_stats['мода'].append(statistics.mode(hr_var[hr_var.columns[i]]))
    except:
        dict_stats['мода'].append('no unique mode')
    i += 1
```

```python
statistics_of_variables = pd.DataFrame(dict_stats)
statistics_of_variables
```

```python
hr_var_left = hr_var[hr_var.left == 1]
hr_var_working = hr_var[hr_var.left == 0]
```

```python
i = 0
dict_stats = {'variables':[], 'среднее': [], 'медиана': [], 'минимум': [], 'максимум': [],
              'стандартное отклонение': [], 'мода': [] }
while i < 8:
    dict_stats['variables'].append(hr_var_working.columns[i])
    dict_stats['среднее'].append(statistics.mean(hr_var_working[hr_var_working.columns[i]])) 
    dict_stats['медиана'].append(statistics.median(hr_var_working[hr_var_working.columns[i]]))
    dict_stats['минимум'].append(min(hr_var_working[hr_var_working.columns[i]]))
    dict_stats['максимум'].append(max(hr_var_working[hr_var_working.columns[i]]))
    dict_stats['стандартное отклонение'].append(statistics.pstdev(hr_var_working[hr_var_working.columns[i]]))
    try:
        dict_stats['мода'].append(statistics.mode(hr_var_working[hr_var_working.columns[i]]))
    except:
        dict_stats['мода'].append('no unique mode')
    i += 1
statistics_of_variables_1 = pd.DataFrame(dict_stats)
statistics_of_variables_1
```

```python
i = 0
dict_stats = {'variables':[], 'среднее': [], 'медиана': [], 'минимум': [], 'максимум': [], 'стандартное отклонение': [], 'мода': []}
while i < 8:
    dict_stats['variables'].append(hr_var_left.columns[i])
    dict_stats['среднее'].append(statistics.mean(hr_var_left[hr_var_left.columns[i]])) 
    dict_stats['медиана'].append(statistics.median(hr_var_left[hr_var_left.columns[i]]))
    dict_stats['минимум'].append(min(hr_var_left[hr_var_left.columns[i]]))
    dict_stats['максимум'].append(max(hr_var_left[hr_var_left.columns[i]]))
    dict_stats['стандартное отклонение'].append(statistics.pstdev(hr_var_left[hr_var_left.columns[i]]))
    try:
        dict_stats['мода'].append(statistics.mode(hr_var_left[hr_var_left.columns[i]]))
    except:
        dict_stats['мода'].append('no unique mode')
    i += 1
statistics_of_variables_0 = pd.DataFrame(dict_stats)
statistics_of_variables_0
```

**5. С использованием t-test статистики проверить взаимосвязи между несколькими переменными.**

Протестируем гипотезу, что руководство получает зарплату выше, чем все прочие департаменты. Нулевая гипотеза: выборки схожи, нет разницы между ними.

```python
management = hr[hr.sales == 'management'][['salary_cat']]
others = hr[hr.sales != 'management'][['salary_cat']]
```

```python
management.salary_cat.value_counts()
```

```python
plt.figure(figsize = [15, 10])
sns.distplot(management)
sns.distplot(others)
plt.legend(['management', 'others'])
plt.xlabel('salary categories: 1-low; 2-middle, 3-high')
plt.show()
```

```python
others.salary_cat.value_counts()
```

```python
print(statistics.pstdev(management.salary_cat)**2, statistics.pstdev(others.salary_cat)**2)
#похожие стандартыне отклонения
```

```python
stats.ttest_ind(management.salary_cat, others.salary_cat, equal_var=False)
```

pvalue почти равно 0, соответственно, мы не можем сказать, что выборки равны, хотя кол-во наблюдений в двух выборках сильно различно. Сравним схожие по размеру выборки.


```python
hr.sales.value_counts()
```

сравним зарплаты менеджеров и маркетологов. Нулевая гипотеза: выборки равны.

```python
marketing = hr[hr.sales == 'marketing'][['salary_cat']]
```

```python
marketing.describe().transpose()
```

```python
management.describe().transpose()
```

```python
plt.figure(figsize = [15, 10])
sns.distplot(management)
sns.distplot(marketing)
plt.legend(['management', 'marketing'])
plt.xlabel('salary categories: 1-low; 2-middle, 3-high')
plt.show()
```

```python
print(statistics.pstdev(management.salary_cat)**2, statistics.pstdev(marketing.salary_cat)**2)
#похожие стандартыне отклонения
```

```python
stats.ttest_ind(management.salary_cat, marketing.salary_cat, equal_var=False)
```

выводы похожи на предыдущий тест: pvalue почти равно 0, соответственно, мы не можем сказать, что выборки равны, хотя кол-во наблюдений в двух выборках сильно различно. Сравним схожие по размеру выборки.

**10. Проверить несколько гипотез для данных, чтобы понять почему сотрудники покидают компанию.**

Проверим, что чем ниже зарплата, тем чаще уходят люди.

```python
hrbysalary = hr.groupby(['salary_cat', 'left']).count()[['sales']].rename(columns={'sales': 'number'})
hrbysalary.reset_index(inplace=True)
hrbysalary
```

```python
a = 0
percent = []
for i in hrbysalary.number:
    if a % 2 == 0:
        percent.append(i/(i+hrbysalary.number[a+1]))
    else:
        percent.append(i/(i+hrbysalary.number[a-1]))
    a += 1
```

```python
hrbysalary['percent'] = percent
```

```python
hrbysalary = hrbysalary[hrbysalary.left == 1][['salary_cat', 'percent']]
```

```python
plt.figure(figsize=[10, 6])
plt.scatter(hrbysalary['percent'], hrbysalary.salary_cat)
plt.xlabel('Percent of employees left')
plt.ylabel('Salary category: 1-low, 3-high')
plt.title('Graphical representation of dependencies of % of those who left the company in every salary category')
plt.show()
```

```python
hrbysalary.corr()
```

И на графике видно, и коэффициент корреляции показывает, что есть высокий уровень зависимости. Вопрос в том, является ли  эта зависимость причинно-следственной связью. Логика подсказывает, что сотрудники с более маленькой зарплатой более склонны менять работу (это и проще, и вакансий больше). Тем не менее именно причиной ухода вряд ли является зарплата.
